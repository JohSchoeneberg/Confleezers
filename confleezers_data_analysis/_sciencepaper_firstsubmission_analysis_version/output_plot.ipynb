```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from numpy import genfromtxt
```

```python
xlimit = (-180,200)

c_ch1 = 'cyan'
c_ch2 = 'lime'
c_ch3 = 'red'
```

```python
tube_force = genfromtxt('./output/out_tube_force.csv', delimiter=',')
tube_intensity = genfromtxt('./output/out_tube_fluorescenceIntensity.csv', delimiter=',')
tube_intensity_corrected_ch1 = genfromtxt('./output/out_tube_fluorescenceIntensity_ch1_corrected.csv', delimiter=',')
tube_intensity_corrected_ch2 = genfromtxt('./output/out_tube_fluorescenceIntensity_ch2_corrected.csv', delimiter=',')
tube_intensity_corrected_ch3 = genfromtxt('./output/out_tube_fluorescenceIntensity_ch3_corrected.csv', delimiter=',')
tube_length = genfromtxt('./output/out_tube_length.csv', delimiter=',')
tube_fwhm = genfromtxt('./output/out_tube_tubeRadius_FWHM.csv', delimiter=',')

#vesicle_intensity_corrected_ch1 = genfromtxt('out_vesicle_fluorescenceIntensity_ch1_corrected.csv', delimiter=',')
vesicle_intensity_corrected_ch2 = genfromtxt('./output/out_vesicle_fluorescenceIntensity_ch2_corrected.csv', delimiter=',')
vesicle_intensity_corrected_ch1 = vesicle_intensity_corrected_ch2
vesicle_intensity_corrected_ch3 = genfromtxt('./output/out_vesicle_fluorescenceIntensity_ch3_corrected.csv', delimiter=',')
```

```python
fig = plt.figure(dpi=150)

ax1 = fig.add_subplot(221)
ax1.plot(tube_force[:,0],tube_force[:,1],c='k',lw=2)
#ax1.set_xlabel("time [s]")
ax1.set_ylabel("force [pN]")
ax1.axvline(0, color='k', linestyle='--', lw=0.8)
ax1.set_xlim(xlimit[0],xlimit[1])

ax2 = fig.add_subplot(222)
#ch1
#idx1 = 190
#ch1time = (tube_intensity_corrected_ch1[:,0]+60)[idx1:-1]
#ch1mean = (tube_intensity_corrected_ch1[:,1]/np.average(tube_intensity_corrected_ch1[:,1][200:210]))[idx1:-1]
#ch1std = (tube_intensity_corrected_ch1[:,2])[idx1:-1]
#ax2.plot(ch1time,ch1mean,c=c_ch1,lw=2)
#ax2.fill_between(ch1time,ch1mean-1*ch1std*ch1mean, ch1mean+1*ch1std*ch1mean, color=c_ch1, alpha=0.2)

idx1 = 30
ch1time = (tube_intensity_corrected_ch1[:,0]+147)[idx1:-1]
ch1mean = (tube_intensity_corrected_ch1[:,1]/np.average(tube_intensity_corrected_ch1[:,1][100:150]))[idx1:-1]
ch1std = (tube_intensity_corrected_ch1[:,2])[idx1:-1]/2
ax2.plot(ch1time,ch1mean,c=c_ch1,lw=2)
ax2.fill_between(ch1time,ch1mean-1*ch1std*ch1mean, ch1mean+1*ch1std*ch1mean, color=c_ch1, alpha=0.2)

#ch2
ch2time = tube_intensity_corrected_ch2[:,0]-10
ch2mean = tube_intensity_corrected_ch2[:,1]/np.average(tube_intensity_corrected_ch2[:,1][0:10])
ch2std = tube_intensity_corrected_ch2[:,2]
ax2.plot(ch2time,ch2mean,c=c_ch2,lw=2)
ax2.fill_between(ch2time,ch2mean-1*ch2std*ch2mean, ch2mean+1*ch2std*ch2mean, color=c_ch2, alpha=0.2)


#ch3
ch3time = tube_intensity_corrected_ch3[:,0]+57
ch3mean = tube_intensity_corrected_ch3[:,1]/np.average(tube_intensity_corrected_ch3[:,1][0:10])
ch3std = tube_intensity_corrected_ch3[:,2]
ax2.plot(ch3time,ch3mean,c=c_ch3,lw=2)
ax2.fill_between(ch3time,ch3mean-1*ch3std*ch3mean, ch3mean+1*ch3std*ch3mean, color=c_ch3, alpha=0.2)



ax2.set_ylabel("tube intensity [AU]")
ax2.axvline(0, color='k', linestyle='--', lw=0.8)
ax2.set_xlim(xlimit[0],xlimit[1])

ax3 = fig.add_subplot(223)
ax3.plot(tube_length[:,0],tube_length[:,1],c='k',lw=2)
ax3.set_xlabel("time [s]")
ax3.set_ylabel("length [¬µm]")
ax3.axvline(0, color='k', linestyle='--', lw=0.8)
ax3.set_xlim(xlimit[0],xlimit[1])


ax4 = fig.add_subplot(224)


#ch1 vesicle
chtime = vesicle_intensity_corrected_ch1[:,0]
chmean = vesicle_intensity_corrected_ch1[:,1]/np.average(vesicle_intensity_corrected_ch1[:,1][0:10])
chstd = vesicle_intensity_corrected_ch1[:,2]
ax4.plot(chtime,chmean,c=c_ch1,lw=2)
ax4.fill_between(chtime,chmean-1*chstd*chmean, chmean+1*chstd*chmean, color=c_ch1, alpha=0.2)

#ax4.plot(vesicle_intensity_corrected_ch1[:,0],vesicle_intensity_corrected_ch1[:,1]/np.average(vesicle_intensity_corrected_ch1[:,1][0:10]),c=c_ch1,lw=2)

#ch2 vesicle
chtime = vesicle_intensity_corrected_ch2[:,0]
chmean = vesicle_intensity_corrected_ch2[:,1]/np.average(vesicle_intensity_corrected_ch2[:,1][0:10])
chstd = vesicle_intensity_corrected_ch2[:,2]
ax4.plot(chtime,chmean,c=c_ch2,lw=2)
ax4.fill_between(chtime,chmean-1*chstd*chmean, chmean+1*chstd*chmean, color=c_ch2, alpha=0.2)
#ax4.plot(vesicle_intensity_corrected_ch2[:,0],vesicle_intensity_corrected_ch2[:,1]/np.average(vesicle_intensity_corrected_ch2[:,1][0:10]),c=c_ch2,lw=2)

#ch3 vesicle
ch3time = vesicle_intensity_corrected_ch3[:,0]-17
ch3mean = vesicle_intensity_corrected_ch3[:,1]/np.average(vesicle_intensity_corrected_ch3[:,1][0:10])
ch3std = vesicle_intensity_corrected_ch3[:,2]
ax4.plot(ch3time,ch3mean,c=c_ch3,lw=2)
ax4.fill_between(ch3time,ch3mean-1*ch3std*ch3mean, ch3mean+1*ch3std*ch3mean, color=c_ch3, alpha=0.2)

ax4.set_ylabel("vesicle intensity [AU]")
ax4.set_xlabel("time [s]")
ax4.axvline(0, color='k', linestyle='--', lw=0.8)
ax4.set_xlim(xlimit[0],xlimit[1])


plt.tight_layout()
fig = plt.gcf()

plt.savefig("out_plot.png")
```
